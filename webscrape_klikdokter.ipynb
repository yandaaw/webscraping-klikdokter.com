```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
from lxml import html
import string
```

### Link

```python
#user agent = browser yang akses linknya
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0'}
#request dapetin isinya (ke page obat)
#text = requests.get("https://www.klikdokter.com/obat", headers=headers).text
```

```python
#create a string of all uppercase letters
alphabet_string = string.ascii_lowercase

#create list
alphabet_list = list(alphabet_string)

for a in alphabet_list:
    #bikin array kosong buat store link
        links = []
        
        for b in alphabet_list:
            #request dapetin isinya 
            link = requests.get('https://www.klikdokter.com/obat?alphabet=' + b + '&partial=1', headers=headers).text
            #???
            soup = BeautifulSoup(link, 'lxml')
            #???
            g_data = soup.find_all("a", {"class": "topics-index--section__item-link"})

            #ambil href = hypertext reference
            for item in g_data:
                data = item["href"]
                links.append(data)
            
        links.append(g_data)
```

```python
links
```

```python
#bikin array kosong buat description
desc = []


for x in links:
    try: 
        text = requests.get(str(x), headers=headers) #ambil konten dari setiap link
        tree = html.fromstring(text.content) #buat buka konten inspect
        elements = tree.xpath('//*[@id="main"]/div[2]/div/div/div[2]/div/p//text()') #cari yg id=main terus buka div ke2, div ke1 nya yg ke2, dst
                                                                                        #pake // biar semua konten di dalam p[1] ke ambil semua, tanpa peduli ada blocker atau enggak
                                                                                        #klo pake / , jika ada yg di dalam directory lain kyk <em>...</em>, akan dilewati
        elements = pd.DataFrame(elements)[0] #???
        elements['links'] = x
        desc.append(elements) #array diisi element
    except:
        pass
```

```python
#bikin dataframe
df = pd.DataFrame(desc)
```
