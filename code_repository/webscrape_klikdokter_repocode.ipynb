```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
from lxml import html
```

### Link

```python
#user agent = browser yang akses linknya
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0'}
#request dapetin isinya (ke page obat)
#text = requests.get("https://www.klikdokter.com/obat", headers=headers).text
```

```python
#request dapetin isinya (ke page obat abjad e)
link = requests.get('https://www.klikdokter.com/obat?alphabet=e&partial=1', headers=headers).text
#???
soup = BeautifulSoup(link, 'lxml')

#???
g_data = soup.find_all("a", {"class": "topics-index--section__item-link"})

#bikin array kosong buat store link
links = []

#ambil href = hypertext reference
for item in g_data:
    data = item["href"]
    links.append(data)
```

```python
links
```

```python
#bikin array kosong buat description
desc = []


for x in links:
    try: 
        text = requests.get(str(x), headers=headers) #ambil konten dari setiap link
        tree = html.fromstring(text.content) #buat buka konten inspect
        elements = tree.xpath('//*[@id="main"]/div[2]/div/div/div[2]/div/p//text()') #cari yg id=main terus buka div ke2, div ke1 nya yg ke2, dst
                                                                                        #pake // biar semua konten di dalam p[1] ke ambil semua, tanpa peduli ada blocker atau enggak
                                                                                        #klo pake / , jika ada yg di dalam directory lain kyk <em>...</em>, akan dilewati
        elements = pd.DataFrame(elements)[0] #???
        elements['links'] = x
        desc.append(elements) #array diisi element
    except:
        pass
```

```python
#bikin dataframe
df = pd.DataFrame(desc)
```

```python
df
```

```python
#pindahin link ke depan, ambil namanya
#ambil nama obat dari link
df['links'] = df['links'].str.split("obat/").str[1]

# shift column 'link' to first position
first_column = df.pop('links')
df.insert(0, 'product_name', first_column)

#gabungin kolom-kolom deskripsi
#menggabungkan semua kolom dari kolom index ke 1 + sebelum join ngedrop kolom yg null dan diubah jadi str, axis 
df['description'] = df[df.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1 #ngeliat secara kolom (horizontal), mendukung penggabungan kolom
)

#hapus teks dari 'Artikel Lainnya' ke belakang
df['description'] = df['description'].str.split("Artikel Lainnya").str[0]

#hapus teks dari 'Cara Penyimpanan' ke belakang
df['description'] = df['description'].str.split("Cara Penyimpanan").str[0]

#hapus teks dari 'Cara Penggunaan' ke belakang
df['description'] = df['description'].str.split("Cara Penggunaan").str[0]

#hapus teks dari 'Aturan Penggunaan' ke belakang
df['description'] = df['description'].str.split("Aturan Penggunaan").str[0]

#hapus teks dari 'Kontraindikasi' ke belakang
df['description'] = df['description'].str.split("Kontraindikasi").str[0]

#hapus teks dari 'Dosis pemberian:' ke belakang
df['description'] = df['description'].str.split("Dosis pemberian:").str[0]
```

```python
#bikin dataframe isi 2 kolom
df = df[['product_name', 'description']]

# export
#df.to_excel("crawling_test_4.xlsx", index=False, header=True)
```
